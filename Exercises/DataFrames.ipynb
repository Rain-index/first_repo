**Series и DataFrame**
Данные в Pandas представлены в двух видах: Series и DataFrame. Разберёмся с каждым из них.

**Series** — это объект, который похож на одномерный массив и может содержать любые типы данных. Проще всего представить его как столбец таблицы, с последовательностью каких-либо значений, у каждого из которых есть индекс — номер строки.

Создадим простой Series:

```python
# Импортируем библиотеку Pandas
import pandas as pd
# Создаём объект Series, содержащий числа
series_example = pd.Series([4, 7, -5, 3])
# Выводим объект на экран
series_example
```

Series отображается в виде таблицы с индексами элементов в первом столбце и значениями во втором.

**DataFrame** — основной тип данных в Pandas, вокруг которого строится вся работа. Его можно представить в виде обычной таблицы с любым количеством столбцов и строк. Внутри ячеек такой «таблицы» могут быть данные самого разного типа: числовые, булевы, строковые и так далее.

У **DataFrame** есть и индексы строк, и индексы столбцов. Это позволяет удобно сортировать и фильтровать данные, а также быстро находить нужные ячейки.

Создадим простой DataFrame с помощью словаря и посмотрим на его отображение:

```python
# Создаём словарь с нужной информацией о городах
city = {'Город': ['Москва', 'Санкт-Петербург', 'Новосибирск', 'Екатеринбург'],
        'Год основания': [1147, 1703, 1893, 1723],
        'Население': [11.9, 4.9, 1.5, 1.4]}
# Превращаем словарь в DataFrame, используя стандартный метод библиотеки
df = pd.DataFrame(city)
# Выводим DataFrame на экран
df
```

Мы видим таблицу, строки которой имеют индексы от 0 до 3, а «индексы» столбцов соответствуют их названиям. Легко заметить, что DataFrame состоит из трёх Series: «Город», «Год основания» и «Население». Оба типа индексов можно использовать для навигации по данным.

**Упражнение 3: Построение и анализ гистограмм**
В этом упражнении мы создадим гистограмму частоты алмазов в наборе данных с указанием их соответствующих характеристик в каратах по оси X:

```python
# importing the necessary libraries
import seaborn as sns
import pandas as pd
```

```python
# Загрузим еще раз данные, без дополнительных столбцов
diamonds_df = sns.load_dataset('diamonds')
diamonds_df.head()
```

```python
# Построим гистограмму, используя датасет diamonds, где ось х = караты:
diamonds_df.hist(column='carat')
```

Ось Y на этом графике обозначает **количество бриллиантов** в наборе данных со **спецификацией в каратах** по оси X. Функция **hist** имеет параметр, называемый **bins (интервалами)**, который буквально означает количество интервалов одинакового размера, на которые разделены точки данных. По умолчанию параметр bins в pandas установлен на 10. Если захотим, мы можем изменить это число на другое.

```python
# Изменим параметр bins = 50:
diamonds_df.hist(column='carat', bins=50)
```

Это гистограмма с **50 интервалами**. Обратите внимание - мы видим более детальное распределение по мере увеличения количества интервалов.

Теперь давайте посмотрим на ту же функцию, используя **seaborn**:

```python
sns.displot(diamonds_df.carat, kde=True, kde_kws=dict(cut=3), bins=50)
```

Есть два заметных различия между функцией **pandas hist** и **seaborn distplot**:

• **pandas** устанавливает для параметра bins значение **по умолчанию, равное 10**, но **seaborn** определяет подходящий размер ячейки на основе **статистического распределения** набора данных.

• По умолчанию функция **distplot** также включает сглаженную кривую на гистограмме, называемую **оценкой плотности ядра**.

**Оценка плотности ядра (KDE)** — это непараметрический способ оценки функции плотности вероятности случайной величины. Обычно KDE не сообщает нам ничего, кроме того, что мы можем вывести из самой гистограммы. Однако это полезно при сравнении нескольких гистограмм на одном графике.

Если мы хотим удалить KDE и просмотреть только гистограмму, мы можем использовать параметр **kde=False**.

```python
# Change kde=False to remove the KDE:
sns.displot(diamonds_df.carat, kde=False, kde_kws=dict(cut=3), bins=50)
```

Также обратите внимание, что параметр bins отображает более подробный график. Теперь давайте попробуем увеличить его до 100.

```python
# Increase the bins size to 100:
sns.displot(diamonds_df.carat, kde=False, kde_kws=dict(cut=3), bins=100)
```

Гистограмма с интервалами **100** демонстрирует лучшую визуализацию распределения переменной — мы видим несколько пиков при определенных значениях карата.

Еще одно наблюдение заключается в том, что большинство значений карата сосредоточены в сторону меньших значений, а хвост находится справа, другими словами, он перекошен вправо.

**Логарифмическое преобразование** помогает выявить больше тенденций. Например, на следующем графике ось X показывает логарифмически преобразованные значения переменной **цена**, и мы видим, что есть два пика, обозначающие два вида бриллиантов — один с высокой ценой, а другой с низкой ценой.

```python
import numpy as np
sns.displot(np.log(diamonds_df.price), kde=False)
```

Это довольно удобно. Глядя на гистограмму, даже наивный зритель сразу получает представление о распределении признака. В частности, в гистограмме важны три наблюдения:

• Какие значения признаков чаще встречаются в наборе данных (в данном случае есть пик около 6,8 и еще один пик между 8,5 и 9 — обратите внимание, что log(price) = ценности в данном случае.

• Сколько пиков существует в данных (пики необходимо дополнительно проверить на предмет возможные причины в контексте данных).

• Есть ли в данных какие-либо выбросы.


**Упражнение 4. Создание гистограммы и расчет распределения средней цены.**
В этом упражнении мы научимся создавать таблицу с помощью функции pandas **crosstab** pandas. Мы будем использовать таблицу для создания гистограммы. Затем мы изучим гистограмму, созданную с помощью библиотеки **seaborn**, и рассчитаем распределение средней цены. Для этого проделаем следующие шаги.

Чтобы получить количество бриллиантов каждого качества огранки, мы сначала создаем таблицу, используя функцию pandas crosstab().

```python
cut_count_table = pd.crosstab(index=diamonds_df['cut'],columns='count')
cut_count_table
```

```python
# Передаем эти значения другой функции pandas, plot(kind='bar')
cut_count_table.plot(kind='bar')
```

```python
# Построим такую же диаграмму, используя seaborn
sns.catplot(data = diamonds_df, x="cut", aspect=1.5, kind="count", color="b")
```

Обратите внимание, что функция **catplot()** не требует от нас создания промежуточной таблицы счетчиков (с использованием pd.crosstab()) и сокращает процесс построения графика на один шаг.

Далее, вот как мы получаем среднее распределение цен на огранку различного качества с использованием Seaborn:

```python
from numpy import median, mean

sns.set(style="whitegrid")
ax = sns.barplot(x="cut", y="price", data=diamonds_df, estimator=mean)
```

Здесь черные линии (полосы погрешностей) на прямоугольниках указывают на неопределенность (или разброс значений) вокруг средней оценки. По умолчанию для этого значения установлена вероятность 95 %.

В вызове ax.legend() параметр ncol обозначает количество столбцов, в которых должны быть организованы значения легенды, а параметр loc указывает расположение легенды и может принимать любое из восьми значений (upper left, lower center и т. д.).

```python
# Аналогичным образом измененим размер шрифта и поворот подписей по оси X:
ax = sns.barplot(x='cut', y='price', hue='color', data = diamonds_df)
ax.legend(loc='upper right',ncol=4)

# установим размер шрифта и вращение меток делений по оси X
ax.set_xticklabels(ax.get_xticklabels(), fontsize=13, rotation=30)
```

**Создание статической диаграммы рассеяния**

Давайте рассмотрим график для представления глобальных закономерностей в данных. Большинство исследователей данных предпочитают видеть такие графики, потому что они дают представление обо всем спектре значений, принимаемых интересующими объектами. Графики, изображающие глобальные закономерности, также полезны, поскольку они облегчают выявление аномалий в данных.
Мы будем работать с набором данных под названием **mpg**. Он был опубликован библиотекой StatLib, поддерживаемой Университетом Карнеги-Меллон, и доступен в библиотеке **seaborn**. Первоначально он использовался для изучения связи пробега (миль на галлон (MPG)) с другими параметрами в наборе данных; отсюда и название миль на галлон. Поскольку набор данных содержит 3 дискретных объекта и 5 непрерывных объектов, он хорошо подходит для иллюстрации нескольких концепций при визуализации.

Вы можете увидеть, как выглядит набор данных, используя:

```python
mpg_df = sns.load_dataset("mpg")
mpg_df.head()
```

Первый тип графика, который мы создадим, — это **диаграмма рассеяния**. **Диаграмма рассеяния** — это простой график, представляющий значения двух объектов в наборе данных. Каждая точка данных представлена точкой с координатой *x* в качестве значения первого признака и координатой *y* в качестве значения *второго признака.

**Диаграмма рассеяния** — отличный инструмент для получения дополнительной информации о двух таких числовых атрибутах.

```python
ax = sns.scatterplot(x="weight", y="mpg", data=mpg_df)
```

Обратите внимание, что диаграмма рассеяния показывает снижение пробега (миль на галлон) с увеличением веса.

Существует также более интересная версия диаграмм рассеяния, называемая **hexagonal binning plot (hexbin plot)** – ее можно использовать, когда и строки, и столбцы соответствуют числовым атрибутам.

При наличии большого количества точек данных точки, нанесенные на диаграмму рассеяния, могут в конечном итоге перекрываться, что приводит к беспорядочному графику. В таких случаях бывает сложно определить тенденции. С помощью шестнадцатеричного графика множество точек данных в одной области можно отобразить более темным оттенком. На графиках Hexbin шестиугольники используются для представления кластеров точек данных. Более темные ячейки указывают на то, что в соответствующих диапазонах объектов по осям x и y имеется большее количество точек. Более светлые ячейки обозначают меньшее количество баллов. Пробелы не соответствуют никаким точкам. Таким образом, мы получаем более чистый и понятный график.

```python
sns.jointplot(data = mpg_df, x = 'weight', y = 'mpg', kind="hex")
```

Как вы могли заметить, гистограмма на верхней и правой осях отображает разницу в характеристиках, представленных осями x и y соответственно (миль на галлон и вес в этом примере). Кроме того, вы могли заметить на предыдущей диаграмме рассеяния, что точки данных сильно перекрываются в определенных областях, скрывая фактическое распределение объектов. Графики Hexbin — довольно хороший инструмент визуализации данных, когда точки данных очень плотные.

Другой альтернативой диаграммам рассеяния, когда точки данных плотно лежат в определенных регионах, является **контурная диаграмма (contour plot)**.

Преимущество использования контурных графиков такое же, как и у шестнадцатеричных графиков: они точно отображают распределение объектов в визуализации в тех случаях, когда точки данных могут сильно перекрываться.

Контурные графики обычно используются для отображения распределения таких показателей погоды, как температура, количество осадков и других, на картах географических регионов.

```python
sns.kdeplot(data = mpg_df, x = 'weight', y = 'mpg', fill=True, alpha=.5)
```

Обратите внимание, что интерпретация контурных графиков аналогична интерпретации шестнадцатеричных графиков: более темные области указывают на большее количество точек данных, а более светлые области указывают на меньшее количество точек данных.

В нашем примере зависимости веса от пробега (миль на галлон) шестигранный график и контурный график показывают, что существует определенная кривая, вдоль которой отрицательная связь между весом и пробегом является наиболее сильной, о чем свидетельствует большее количество точек данных.

Отрицательная связь становится относительно слабее по мере удаления от кривой (меньше точек данных).
